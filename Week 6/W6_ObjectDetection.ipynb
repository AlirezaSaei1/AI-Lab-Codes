# Object Detection: Ball Bounce

### Import required libraries

```python
import cv2
import matplotlib.pyplot as plt
```

### Single Frame

```python
img = cv2.imread('Media\SingleFrame.png')
img = img[:,:,::-1] # Convert RGB to BGR
ycrcb = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
```

```python
# in this channel red balls can be detected easier
plt.imshow(ycrcb[:,:,1])
```

```python
cr_channel = ycrcb[:,:,1]
_, red_mask = cv2.threshold(cr_channel, 150, 255, cv2.THRESH_BINARY)
plt.imshow(red_mask)
```

```python
contours, _ = cv2.findContours(red_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

img_with_rectangles = img.copy()

for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(img_with_rectangles, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
plt.imshow(img_with_rectangles)
plt.axis('off')
plt.show()
```
