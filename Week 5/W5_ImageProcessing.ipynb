### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Read Image + Color Spaces

```python
img = cv2.imread('Media\Statue_of_Liberty.jpg')
img = img[:,:,::-1] # Convert RGB to BGR
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) # Convert RGB to Gray Scale (Calculates Mean of RGB)
ycrbr = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(img)
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(gray, cmap='gray')
axs[1].set_title('Grayscale Image')
axs[1].axis('off')

axs[2].imshow(ycrbr)
axs[2].set_title('YCrCb Image')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```

### Threshold

```python
ret, thresh = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
merged_img = cv2.bitwise_and(img, img, mask=thresh)
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].imshow(thresh, cmap='gray')
axs[0].set_title('Thresholded Image')
axs[0].axis('off')

axs[1].imshow(merged_img, cmap='gray')
axs[1].set_title('Thresholded Original Image')
axs[1].axis('off')

plt.show()
```

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
plt.imshow(laplacian.astype(np.uint8), cmap='gray')
```
